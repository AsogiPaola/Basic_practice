Suposiciones que se deben hacer antes de realizar la regresion lineal para asegurarse de que todo esta en orden:
* Relacion lineal
* normalidad
* no multicolinealidad
* no auto correlacion
* varianzas iguales

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv("data/ingreso.csv")
data
```

```python
plt.ylabel("Ingreso ($MX)")
plt.xlabel("Promedio de horas semanales trabajadas")
plt.scatter(data["horas"],  data["ingreso"], color = "pink")
```

```python
regresion = linear_model.LinearRegression()

#preparar los datos
#data["horas"]
horas = data["horas"].values.reshape((-1,1))
model = regresion.fit(horas, data["ingreso"])

print("Interseccion (b)", model.intercept_)
print("Pendiente (m)", model.coef_)
```

```python
#prediciendo el salario para una persona que trabaja 39, 40, 43 o 44 horas a la semana
entrada = [[39],[40],[43],[44]]
model.predict(entrada)
```

```python
plt.scatter(entrada, model.predict(entrada), color= "red")
plt.plot(entrada, model.predict(entrada), color="blue")

plt.ylabel("Ingreso ($MX)")
plt.xlabel("Promedio de horas semanales trabajadas")
plt.scatter(data["horas"],  data["ingreso"], color = "pink")
```
